```python
import gzip
import ast

path = "meta_Movies_and_TV.json.gz"

print("开始读取电影元数据...\n")
success = 0

# 【终极容错】二进制读取 + latin-1解码 + 跳过所有坏行
with gzip.open(path, "rb") as f:
    for line in f:
        try:
            # 亚马逊meta必须用latin-1解码
            line_str = line.decode("latin-1").strip()
            if not line_str:
                continue
            
            movie = ast.literal_eval(line_str)
            
            # 提取核心信息
            asin = movie.get("asin", "无ID")
            title = movie.get("title", "无标题")
            img = movie.get("imUrl", "无海报")
            desc = movie.get("description", "无简介")
            
            print(f"===== 第 {success+1} 条 =====")
            print(f"物品ID: {asin}")
            print(f"标题: {title}")
            print(f"海报URL: {img}")
            print(f"简介: {desc[:120]}...\n")
            
            success += 1
            if success >= 5:  # 只打印5条
                break
            
        except:
            # 跳过所有脏数据/坏行
            continue

print(f"✅ 成功读取 {success} 条电影数据！")
```

```python
import gzip
import ast
import os

def read_amazon_gz(file_name, data_type):
    """
    读取亚马逊.gz压缩数据
    file_name: 文件名
    data_type: meta(元数据) / review(交互数据)
    """
    if not os.path.exists(file_name):
        print(f"❌ 未找到文件：{file_name}\n")
        return

    print(f"="*60)
    print(f"📂 正在读取：{file_name}")
    print(f"📌 数据类型：{'元数据(标题/海报/简介)' if data_type=='meta' else '交互数据(用户时序行为)'}")
    print(f"="*60)

    total_count = 0  # 总有效数据量
    show_count = 0   # 打印前5条

    # 二进制读取压缩包 + latin-1编码（亚马逊标准）
    with gzip.open(file_name, 'rb') as f:
        for line in f:
            try:
                # 解码 + 解析Python字典（兼容单引号/双引号）
                line_str = line.decode('latin-1').strip()
                if not line_str:
                    continue
                data = ast.literal_eval(line_str)
                total_count += 1

                # 只打印前5条
                if show_count < 5:
                    print(f"\n🔹 第 {show_count+1} 条数据")
                    if data_type == "meta":
                        # 元数据：电影/图书 海报、标题、ID
                        print(f"物品ID: {data.get('asin', '无')}")
                        print(f"标题: {data.get('title', '无')[:80]}...")
                        print(f"图片链接: {data.get('imUrl', '无')}")
                        print(f"简介: {data.get('description', '无')[:100]}...")
                    else:
                        # 交互数据：用户ID、物品ID、时间戳、评分
                        print(f"用户ID: {data.get('reviewerID', '无')}")
                        print(f"物品ID: {data.get('asin', '无')}")
                        print(f"时间戳: {data.get('unixReviewTime', '无')}")
                        print(f"评分: {data.get('overall', '无')}")
                    show_count += 1

            except:
                # 自动跳过脏数据/错误行
                continue

    print(f"\n✅ {file_name} 读取完成！")
    print(f"📊 总有效数据量：{total_count} 条\n\n")

if __name__ == "__main__":
    # 1. 电影元数据
    read_amazon_gz("meta_Movies_and_TV.json.gz", data_type="meta")
    
    # 2. 电影交互数据
    read_amazon_gz("reviews_Movies_and_TV.json.gz", data_type="review")
    
    # 3. 图书元数据
    read_amazon_gz("meta_Books.json.gz", data_type="meta")
    
    # 4. 图书交互数据
    read_amazon_gz("reviews_Books.json.gz", data_type="review")
```

```python
import gzip
import ast
import os
from pprint import pprint  # 格式化打印字典，清晰展示所有字段

def read_amazon_gz(file_name, data_type):
    """
    读取亚马逊.gz压缩数据，输出前5条【完整原始数据】，查看数据格式
    file_name: 文件名
    data_type: meta(元数据) / review(交互数据)
    """
    if not os.path.exists(file_name):
        print(f"❌ 未找到文件：{file_name}\n")
        return

    print(f"="*80)
    print(f"📂 正在读取：{file_name}")
    print(f"📌 数据类型：{'元数据(物品信息)' if data_type=='meta' else '交互数据(用户行为)'}")
    print(f"="*80)

    total_count = 0  # 总有效数据量
    show_count = 0   # 打印前5条（完整数据）

    # 二进制读取压缩包 + latin-1编码（亚马逊数据集标准编码）
    with gzip.open(file_name, 'rb') as f:
        for line in f:
            try:
                line_str = line.decode('latin-1').strip()
                if not line_str:
                    continue
                # 解析为Python字典（原始完整数据）
                data = ast.literal_eval(line_str)
                total_count += 1

                # 打印前5条【所有字段】，用于查看数据格式
                if show_count < 5:
                    print(f"\n🔹 第 {show_count + 1} 条完整数据：")
                    pprint(data)  # 格式化输出所有元素，结构清晰
                    print("-"*80)
                    show_count += 1
                else :
                    break
            except Exception:
                # 跳过异常数据，不中断程序
                continue

    print(f"\n✅ {file_name} 读取完成！")
    print(f"📊 总有效数据量：{total_count} 条\n\n")

if __name__ == "__main__":
    # 依次读取4个文件，每个输出前5条完整数据
    # 1. 电影元数据
    read_amazon_gz("meta_Movies_and_TV.json.gz", data_type="meta")
    
    # 2. 电影交互数据
    read_amazon_gz("reviews_Movies_and_TV.json.gz", data_type="review")
    
    # 3. 图书元数据
    read_amazon_gz("meta_Books.json.gz", data_type="meta")
    
    # 4. 图书交互数据
    read_amazon_gz("reviews_Books.json.gz", data_type="review")
```
